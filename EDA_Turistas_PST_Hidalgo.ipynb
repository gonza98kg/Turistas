Análisis Exploratorio de Datos para el Desarrollo de una Aplicación de Turismo en Hidalgo
</h2>

Se recibió la base de datos resultante de un instrumento de levantamiento de información aplicado en tres destinos turísticos del Estado de Hidalgo, diseñado e implementado por el área correspondiente. El instrumento fue aplicado a dos poblaciones objetivo: **807 turistas** y **aproximadamente 100 Prestadores de Servicios Turísticos (PST)**.

El presente notebook documenta el proceso de limpieza, estandarización y análisis exploratorio de ambas bases de datos. El análisis busca responder preguntas clave sobre el perfil de cada población:

- **Turistas:** ¿De dónde provienen? ¿Cómo y cuándo viajan? ¿Qué buscan en un destino? ¿Qué esperan de una aplicación de turismo desarrollada por el gobierno del estado?
- **PST:** ¿Cuál es su perfil y tipo de servicio? ¿Qué dificultades enfrentan? ¿Qué canales de promoción utilizan? ¿Qué esperan de una plataforma digital?

Los hallazgos e insights derivados del análisis fueron entregados al equipo de desarrollo como insumo para la definición de los MVP de la aplicación. Los datos procesados se exportan como archivos `.csv` limpios para su visualización en **Looker Studio**.

En la columna $"\texttt{Tipo de Servicio que se ofrece}"$ se encuentra que las respuestas pueden agruparse por categorías, por ejemplo, se tiene la categoría "Artesanías" y muchas de las respuestas están como "Otro: Artesanía"

```python
conteo_servicios = df_PST['Tipo de Servicio que se ofrece: '].value_counts().sort_index()
print(conteo_servicios)
```

A partir de la lista anterior se crea una función para clasificar servicios en distintas categorías como "Hospedaje", "Alimentos y Bebidas", "Artesanías y Comercio", "Actividades y Experiencias", "Transporte"

```python
def clasificar_servicio(texto):
    if pd.isna(texto):
        return 'Sin Dato'
    texto = str(texto).lower()
    
    # HOSPEDAJE
    if any(x in texto for x in ['hotel', 'cabaña', 'posada', 'hospedaje', 'airbnb', 'hostal', 'glamping', 'habitacion', 'habitación']):
        return 'Hospedaje'
    
    # ALIMENTOS Y BEBIDAS
    elif any(x in texto for x in ['restaurante', 'comida', 'alimento', 'cocina', 'bar', 'café', 'cafe', 
                                  'pulque', 'pastes', 'gastronom', 'panader', 
                                  'pollo', 'tortilla', 'botana', 'helado', 'dulce', 'fruta', 'verdura']):
        return 'Alimentos y Bebidas'
    
    # 3 EXPERIENCIAS
    elif any(x in texto for x in ['guía', 'guia', 'tour', 'recorrido', 'balneario', 'temazcal', 
                                  'museo', 'parque', 'senderismo', 'experiencia', 'taller', 'zona arque', 'salud']):
        return 'Actividades y Experiencias'
    
    # 4. ARTESANÍAS Y COMERCIO
    elif any(x in texto for x in ['artesan', 'souvenir', 'regalo', 'ixtle', 'bordado', 'tienda', 
                                  'abarrotes', 'miscelanea', 'papeleria', 'farmacia', 'medicina', 
                                  'limpieza', 'ropa', 'tiend', 'internet']):
        return 'Artesanías y Comercio'
    
    # 5. TRANSPORTE
    elif any(x in texto for x in ['transporte', 'taxi', 'traslado', 'renta']):
        return 'Transporte'
        
    else:
        return 'Otros Servicios'

df_PST['Tipo de Servicio que se ofrece: '] = df_PST['Tipo de Servicio que se ofrece: '].apply(clasificar_servicio)
print(df_PST['Tipo de Servicio que se ofrece: '].value_counts())
```

En la columna $\texttt{¿Cómo te das a conocer actualmente?}$ algunas de las respuestas que entran en la categoría de "Redes sociales" se registraron como "Otro" por lo que se homogeneizará en este caso.

```python
columna = '¿Cómo te das a conocer actualmente?'

reemplazos = {
    'Otro: Redes sociales y personalmente': 'Redes sociales',
    'Otro: Google maps': 'Redes sociales'
}

df_PST[columna] = df_PST[columna].replace(reemplazos)

print(df_PST[columna].value_counts())
```

Para evaluar la efectividad de los canales de adquisición en la columna $\texttt{"De esos medios: Elige 3 que realmente te traen clientes"}$, se realizará un desglose de frecuencias de respuesta múltiple sobre la variable 'Medios que traen clientes'. Dado que los prestadores de servicios podían seleccionar varias opciones simultáneamente (ej. 'Redes sociales, Recomendaciones'), se separarán estos registros en valores individuales para cuantificar la incidencia real de cada canal. Este procedimiento permitirá jerarquizar los medios de promoción por su efectividad aislada, revelando si la tracción turística en Hidalgo depende mayoritariamente de esfuerzos digitales o de la reputación tradicional (boca a boca).

```python
columna_medios = 'De esos medios: Elige 3 que realmente te traen clientes.'
medios_sucios = df_PST[columna_medios].dropna().str.split(', ', expand=True).stack()

correcciones = {
    '["Redes sociales"': 'Redes sociales',
    '"Ferias"': 'Ferias',
    '"other"] (otro: Google maps y también instagram)': 'Redes sociales',
    '"other"] (otro: Publicidad)': 'Publicidad Tradicional'
}
conteo_final = medios_sucios.replace(correcciones).value_counts()


df_tabla_final = conteo_final.reset_index()
df_tabla_final.columns = ['Medio de Promoción', 'Frecuencia']


print(df_tabla_final)
```

Con el objetivo de facilitar la lectura e interpretación de los gráficos en el tablero de control, se realizó una recodificación de la variable binaria sobre el uso de aplicaciones digitales. Los valores lógicos originales (True/False) fueron sustituidos por etiquetas categóricas explícitas ('Sí' y 'No'), eliminando la necesidad de traducción técnica por parte de la audiencia final.

```python

columna_apps = '¿Usas alguna app o plataforma para promocionar tu servicio? ¿Cómo ha sido tu experiencia?'

cambios = {
    'FALSO': 'No', 
    'VERDADERO': 'Sí',
}
df_PST.loc[:, columna_apps] = df_PST[columna_apps].replace(cambios)

print(df_PST[columna_apps].value_counts())
```

Dada la naturaleza cualitativa de la variable sobre estacionalidad, se aplicó un proceso de categorización semántica para estructurar las respuestas abiertas. Se generó una nueva variable analítica ('Estacionalidad_Categoria') que clasifica los comentarios en dimensiones medibles: Dependencia de Calendario (vacaciones/festivos) y Vulnerabilidad Económica (impacto en ingresos). Esta transformación permite cuantificar no solo los periodos de baja afluencia, sino la severidad con la que los prestadores de servicios perciben la inestabilidad de la demanda.

```python
col_original = '¿Hay temporadas donde tienes muchos visitantes y otras donde casi no llegan? ¿Cómo te afecta eso?'

def clasificar_estacionalidad(texto):
    if pd.isna(texto): return 'Sin Dato'
    t = str(texto).lower()
    
    if any(x in t for x in ['vacacion', 'vacion', 'semana', 'marzo', 'diciembre', 'navidad', 'puente', 'feria']):
        return 'Temporada Vacacional'
    elif any(x in t for x in ['fin de semana', 'fines', 'sabado', 'domingo']):
        return 'Fines de Semana'
    elif any(x in t for x in ['econom', 'venta', 'ingreso', 'dinero', 'baja', 'afecta', 'poco', 'gente']):
        return 'Impacto Económico / Baja Afluencia'
    elif 'no' in t and len(t) < 15: # Respuestas cortas negativas
         return 'Flujo Constante / No afecta'
    else:
        return 'Variable / Otro'

valores_nuevos = df_PST[col_original].apply(clasificar_estacionalidad)

indice = df_PST.columns.get_loc(col_original)

col_nueva = 'Estacionalidad_Categoria'
if col_nueva in df_PST.columns:
    df_PST[col_nueva] = valores_nuevos 
else:
    df_PST.insert(loc=indice + 1, column=col_nueva, value=valores_nuevos)

print(df_PST[col_nueva].value_counts())
```

Para cuantificar las barreras de crecimiento, se procesó la variable abierta sobre problemáticas principales mediante un análisis de frecuencia de palabras clave. Las respuestas textuales originales fueron sobrescritas y estandarizadas en cuatro categorías macro: Deficiencias de Infraestructura/Acceso (carreteras, señalética), Falta de Visibilidad (promoción, publicidad), Limitaciones Económicas y Otros Factores. Esta recodificación permite identificar claramente el 'cuello de botella' principal que frena la llegada de turistas.

```python
def clasificar_problema(texto):
    if pd.isna(texto): return 'Sin Dato'
    t = str(texto).lower()
    
    if any(x in t for x in ['carretera', 'camino', 'transporte', 'vialidad', 'señal', 'calle', 'acceso', 'bache', 'obra', 'ubicacion', 'llegar']):
        return 'Infraestructura y Acceso'
    elif any(x in t for x in ['publicidad', 'promocion', 'difusion', 'conocen', 'visibilidad', 'redes', 'marketing', 'internet', 'tecnolog']):
        return 'Falta de Publicidad / Visibilidad'
    elif any(x in t for x in ['dinero', 'recurso', 'econom', 'costo', 'presupuesto', 'inver', 'caro']):
        return 'Falta de Recursos / Inversión'
    elif any(x in t for x in ['ningun', 'nada', 'todo bien']):
        return 'Ninguno'
    else:
        return 'Otro Factor / No Especifica'

col_problema = '¿Cuál dirías que es el principal problema que enfrentas para atraer más visitantes?'

df_PST[col_problema] = df_PST[col_problema].apply(clasificar_problema)

print(df_PST[col_problema].value_counts())
```

Se evaluó la barrera de entrada tecnológica mediante la recodificación de la variable de percepción de uso digital. Las respuestas abiertas, que variaban desde negaciones simples hasta explicaciones complejas, fueron homologadas en una escala de tres niveles: 'Fácil / Accesible', 'Regular / Neutro' y 'Difícil / Complicado'. Esta estandarización permite cuantificar el porcentaje de la población objetivo que requiere capacitación técnica prioritaria frente a aquellos que ya dominan las herramientas básicas.

```python
def clasificar_facilidad(texto):
    if pd.isna(texto): return 'Sin Dato'
    t = str(texto).lower()
    
    # DIFÍCIL
    if any(x in t for x in ['dificil', 'difícil', 'complic', 'cuesta', 'batall', 'no le entiendo', 'no se usar', 'no entiendo', 'mucho']):
        # "Mucho" usualmente responde a "¿Qué tan difícil?" -> "Mucho"
        return 'Difícil / Complicado'

    # FÁCIL
    elif any(x in t for x in ['facil', 'fácil', 'sencill', 'no es complic', 'no es dific', 'no me es dif', 'ayuda', 'aprender']):
        return 'Fácil / Accesible'
        
    # REGULAR
    elif any(x in t for x in ['mas o menos', 'más o menos', 'regular', 'poco', 'veces', 'depend']):
        return 'Regular / Neutro'
    
    # Ruido
    else:
        return 'Difícil / Complicado'

col_facilidad = '¿Qué tan fácil o difícil te resulta usar herramientas digitales para promocionarte?\n'

df_PST[col_facilidad] = df_PST[col_facilidad].apply(clasificar_facilidad)
print(df_PST[col_facilidad].value_counts())
```

Para dimensionar las barreras de confianza en el entorno digital, se realizó una clasificación temática de riesgos basada en las experiencias narrativas de los encuestados. La variable original fue estandarizada en tres categorías críticas: Vulnerabilidad Financiera (fraudes y estafas), Gestión de Reputación (malas reseñas/crisis) y Ausencia de Incidencias. Esta transformación permite cuantificar no solo la frecuencia de uso, sino la severidad de los obstáculos reales que frenan la adopción tecnológica en la región.

```python
def clasificar_miedo(texto):
    if pd.isna(texto): return 'Sin Dato'
    t = str(texto).lower()
    
    # 1. FRAUDES
    if any(x in t for x in ['fraude', 'estafa', 'dinero', 'deposito', 'falso', 'robo', 'perfil']):
        return 'Fraudes o Estafas'
        
    # 2. REPUTACIÓN
    elif any(x in t for x in ['reseña', 'comentario', 'disgusto', 'vistas', 'bajan', 'mala']):
        return 'Malas Reseñas / Gestión de Reputación'
    
    # 3. SIN PROBLEMAS
    elif any(x in t for x in ['no', 'ningun', 'nunca', 'todo bien', 'buena']):
        return 'Sin Malas Experiencias'
        
    else:

        return 'Otras Incidencias Técnicas'

col_miedo = '¿Has tenido alguna mala experiencia con plataformas digitales o redes sociales? ¿Qué pasó?'

df_PST[col_miedo] = df_PST[col_miedo].apply(clasificar_miedo)

print(df_PST[col_miedo].value_counts())
```

El análisis de los motivadores de retención revela un enfoque pragmático por parte de los prestadores de servicios turísticos. La continuidad en el uso de la plataforma está condicionada principalmente por la obtención de Resultados Tangibles (aumento directo en ventas y afluencia de visitantes), seguido por la necesidad de Visibilidad y Promoción. Factores como la Facilidad de Uso y el Aprendizaje aparecen como habilitadores secundarios, lo que confirma que la adopción tecnológica en este sector depende menos de la innovación por sí misma y más del retorno de inversión visible que la herramienta pueda generar.

```python
def clasificar_motivacion_final(texto):
    if pd.isna(texto): return 'Sin Dato'
    t = str(texto).lower()
    
    # 1. RESULTADOS ECONÓMICOS
    if any(x in t for x in ['venta', 'ingreso', 'ganancia', 'cliente', 'turista', 'visitante', 'econom', 'resultado', 'beneficio', 'negocio', 'aument']):
        return 'Resultados Tangibles (Ventas/Ingresos)'
        
    # 2. VISIBILIDAD Y ALCANCE
    elif any(x in t for x in ['promocion', 'publicidad', 'conocer', 'visibil', 'difusi', 'vista', 'imagen', 'gente', 'publico', 'atraer']):
        return 'Mayor Visibilidad y Promoción'
        
    # 3. FACILIDAD Y FUNCIONALIDAD
    elif any(x in t for x in ['facil', 'sencill', 'rapid', 'sirv', 'funcion', 'util', 'accesib', 'herramienta', 'manejo']):
        return 'Facilidad de Uso / Utilidad'

    # 4. APRENDIZAJE
    elif any(x in t for x in ['aprender', 'actualiz', 'capacit', 'mejorar', 'superar', 'conocimiento']):
        return 'Aprendizaje y Capacitación'

    # 5. CONFIANZA
    elif any(x in t for x in ['confianza', 'seguridad', 'apoyo', 'oficial', 'gobierno', 'verdad']):
        return 'Confianza y Respaldo Institucional'
        
    else:
        return 'Otros / Indeciso'

col_motiva = '¿Qué te motivaría a seguir usando la plataforma en el tiempo?'
df_PST[col_motiva] = df_PST[col_motiva].apply(clasificar_motivacion_final)

print(df_PST[col_motiva].value_counts())
```
